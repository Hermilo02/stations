# Introducción a Matplotlib

[Matplotlib](https://matplotlib.org) es una librería que permite la creación de figuras y gráficos de calidad mediante el uso de Python.
* Permite la creación de gráficos de manera sencilla y eficiente
* Permite la integración de gráficos y figuras en un Jupyter Notebook

## Import

```python
import matplotlib.pyplot as plt
```

## Representación gráfica de datos

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1,2,5,7,8,3,1])
plt.show()
```

También se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.show()
```

Pueden modificarse las logitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3,-1,0,4,7],[1,4,6,7,8])
plt.axis([-4,8,0,10])#[xmin,xmax,ymin,ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matemática

```python
import numpy as np
```

```python
x=np.linspace(-2,2,1000)
y=x**2

plt.plot(x,y)
plt.show()
```

También pude modificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x,y)
plt.title("Square function $x^2$")
plt.xlabel("x")
plt.ylabel("$y=x^2$")
plt.grid(True)
plt.show()
```

Pueden superponerse gráficas y cambiar el estilo de las líneas

Pueden superponerse gráficas y cambiar el estilo de las líneas
Las opciones disponibles son:

    "-" o "solid": línea sólida (es la opción por defecto)
    "--" o "dashed": línea discontinua 
    "-." o "dashdot": línea que alterna guiones y puntos
    ":" o "dotted": línea de puntos
    "None", " " o "": no muestra nada.

Los colores basicos son:
- "r", rojo      "m", magenta
- "b", azul      "y", amarillo
- "g", verde     "brown", cafe
- "k". negro     "grey", gris
- "w", blanco    "pink", rosa
- "c", cian      "violet", violeta

```python
x=np.linspace(-2,2,1000)
y=x**2
y2=x + 1

plt.plot(x,y,'b-.', x,y2,'r:')
plt.show()
```

```python
# Separando en diferentes lineas las funciones
x=np.linspace(-2,2,1000)
y=x**2
y2=x + 1

plt.plot(x,y,'b-.')
plt.plot(x,y2,'r:')
plt.title("Grafica cuadratica con linea intersectada")
plt.xlabel("Eje X")
plt.ylabel("Eje y")
plt.grid(True)
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda

Location String
Location             Code
- 'best' (Axes only)   0
- 'upper right'        1
- 'upper left'         2
- 'lower left'         3
- 'lower right'        4
- 'right'              5
- 'center left'        6
- 'center right'       7

- 'center'             10

```python
x=np.linspace(-2,2,1000)
y=x**2
y2=x + 1

plt.plot(x,y,'b-.',label="$x^2$")
plt.plot(x,y2,'r:',label="$x+1$")
plt.title("Grafica cuadratica con linea intersectada")
plt.xlabel("Eje X")
plt.ylabel("Eje y")
plt.grid(True)
plt.legend(loc=0)
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan. Estas graficas se organizan en un grid y se denominan subplots.

```python
x=np.linspace(-2,2,1000)
y=x**2
y2=x + 1

plt.subplot(1, 2, 1)# rows, columns, 1st subplot
plt.plot(x,y,'b-.',label="$x^2$")
plt.title("Grafica cuadratica")
plt.xlabel("Eje X1")
plt.ylabel("Eje y1")
plt.grid(True)
plt.legend(loc=0)

plt.subplot(1,2,2)# rows, columns, 2st subplot
plt.plot(x,y2,'r:',label="$x+1$")
plt.title("Grafica lineal")
plt.xlabel("Eje X2")
plt.ylabel("Eje y2")
plt.grid(True)
plt.legend(loc=0)

plt.show()
```

Para que las gráficas no queden tan ajustadas, podemos hacer la figura más grande.

```python
x=np.linspace(-2,2,1000)
y=x**2
y2=x + 1

plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)# rows, columns, 1st subplot
plt.plot(x,y,'b-.',label="$x^2$")
plt.title("Grafica cuadratica")
plt.xlabel("Eje X1", fontsize=12)
plt.ylabel("Eje y1",fontsize=12)
plt.grid(True)
plt.legend(loc=0)

plt.subplot(1,2,2)# rows, columns, 2st subplot
plt.plot(x,y2,'r:',label="$x+1$")
plt.title("Grafica lineal")
plt.xlabel("Eje X2", fontsize=12)
plt.ylabel("Eje y2", fontsize=12)
plt.grid(True)
plt.legend(loc=0)

plt.show()
```

## Scatter plots

```python
from numpy.random import rand
#x=np.random.rand(2,100)
x,y=rand(2,100)

plt.scatter(x,y)
plt.show()
```

```python
x,y=rand(2,100)
x1,y1=rand(2,100)

plt.figure(figsize=(12,6))
plt.subplot(2, 1, 1)# rows, columns, 1st subplot
plt.scatter(x,y,c="red")
plt.subplot(2, 1, 2)# rows, columns, 1st subplot
plt.scatter(x1,y1,c="blue")
plt.show()
```

## Histogramas

```python
data=[1,1.1,1.8,2,2.1,2.2,3.2,3,3,3,3]

plt.figure(figsize=(12,6))
plt.subplot(211)# rows, columns, 1st subplot
plt.hist(data,bins=10,rwidth=0.6)
plt.xlabel("Edad")
plt.ylabel("Frequency")

plt.subplot(212)# rows, columns, 1st subplot
plt.hist(data,bins=[1,1.5,2,2.5,3],rwidth=0.8)
plt.xlabel("Genere")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
